```python
from sklearn import preprocessing
def convert(data):
    number = preprocessing.LabelEncoder()
    for col_name in data.columns:
        data[col_name] = number.fit_transform(data[col_name])
        data=data.fillna(-999)
    return data
```

```python
dataset = convert(ds)
dataset.head(100)
```

```python
X = dataset.drop('class', axis=1)
Y = dataset['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.10)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
```

# Swarn Plot

A swarm plot is a type of categorical scatter plot that displays individual data points along with their distribution. It is particularly useful for visualizing the distribution of categorical data points, especially when there are many categories or a large number of data points within each category.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Age', data=du, inner=None, color='lightgray')
sns.swarmplot(x='BMI', data=du, color='darkblue', alpha=0.5)
plt.title('Age vs BMI Distribution with Swarm Plot')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

# KDE Plot

A KDE (Kernel Density Estimation) plot is a type of data visualization that estimates the probability density function of a continuous variable. It represents the distribution of data in the form of a smooth curve. KDE plots are helpful for visualizing the underlying distribution of a dataset and understanding the data's probability density across different values.

```python
sns.kdeplot(du['BMI'], shade=True, color='orange')
plt.title('BMI vs Age')
plt.xlabel('BMI')
plt.ylabel('Age')
plt.show()
```

# Scatter Plot

A scatter plot is a type of data visualization that displays the relationship between two variables by representing data points on a Cartesian plane. It is commonly used to determine whether there is a correlation or pattern between the two variables.

```python

g = sns.PairGrid(du)
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_lower(sns.kdeplot)
plt.suptitle('classes')
plt.show()
```

# Regression Plot

A regression plot, often created with a linear regression line, is a data visualization tool used to display the relationship between two variables and to show the linear relationship between them. This plot helps in understanding how one variable changes concerning another and is particularly useful for determining the strength and direction of the relationship.

```python

plt.figure(figsize=(8, 6))
sns.regplot(x='Age', y='BMI', data=du)
plt.title('Regression plot of Age vs BMI')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

# Histogram

A histogram is a graphical representation of the distribution of numerical data. It displays the frequencies of values within different intervals or bins. Histograms are useful for understanding the underlying frequency distribution of a continuous variable and identifying patterns or characteristics of the data, such as central tendency, dispersion, and skewness.

```python

# Histogram
sns.histplot(data=du, x='DP Function', bins=20, color='Orange', alpha=0.9, label='DP Function')
sns.histplot(data=du, x='Serum Ins', bins=20, color='Pink', alpha=0.9, label='Serum Ins')
plt.title('Histogram of DP Function and Serum Ins')
plt.xlabel('Serum Ins')
plt.ylabel('DP Function')
plt.legend()
plt.show()



```

# 3-D Scatter Plot

3-D scatter plot visualizes the relationship between three variables by plotting data points in a three-dimensional space.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(du['DP Function'], du['PG Concentration'], du['BMI'])
ax.set_xlabel('BMI')
ax.set_ylabel('DP Function')
ax.set_zlabel('PG Concentration')
plt.title('3D Scatter plot of PG Concentration, DP Function, and BMI')
plt.show()
```
